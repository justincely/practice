```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import scipy
from scipy.stats.stats import pearsonr
from sklearn import ensemble
from sklearn import preprocessing
from sklearn import metrics
```

```python
%matplotlib inline
```

To begin training a model, we'll load the training data into a pandas DataFrame.  After some cleaning and pruning of the data, we'll then proceed to use the scikit-learn framework to create a model.

```python
df = pd.read_table('codetest_2083/codetest_train.txt', delimiter='\t')
```

```python
#-- LabelEncoder will replace strings with unique integers to be used in the classification.
#-- This step is important as the learning algorithms don't work with string parameters.
label_encoder = preprocessing.LabelEncoder()
for colname in df:
    if df[colname].dtype == 'O':
        df[colname] = label_encoder.fit_transform(df[colname])
```

```python
def select_columns(df, min_r=.5e-1):
    """ Select out columns with highest covariance with target sample
    
    Parameters:
    -----------
    df : pandas.DataFrame
        DataFrame containing the target and parameter data
    min_r : float, int
        minimum covariance to specify if a column should be used
        
    Returns:
    --------
    good_columns : list
        the selected columns from the DataFram
        
    Outputs:
    --------
    plot of all 254 parameters vs the target sample
    
    """
    
    good_columns = []

    fig = plt.figure(figsize=(30, 30))
    n_plots = df.shape[1]-1
    n = int(np.sqrt(n_plots)) + 1
    
    for i, colname in enumerate(df):
        if colname == 'target':
            continue

        ax = fig.add_subplot(n, n, i)

        ydata = df['target']
        #-- fill in missing data with 0's. 
        xdata = df[colname].fillna(0)

        r, p = pearsonr(xdata, ydata)
        
        if abs(r) > min_r:
            good_columns.append(colname)
            
            ax.plot(xdata, ydata, 'o', color='b', marker='.')
        else:
            ax.plot(xdata, ydata, 'o', color='grey', marker='.', alpha=.5)

        ax.xaxis.set_ticklabels('')
        ax.yaxis.set_ticklabels('')
        ax.set_title('{}, r:{:3.2e}'.format(i-1, r), color='grey')
        
    fig.suptitle('Parameter covariance with target', fontsize=36)

    fig.savefig('parameters_covariance.png')
    
    return good_columns   
```

```python
good_columns = select_columns(df)
```

The plot above displays all 254 parameters plotted against the targets, with the correlation coefficient displayed above each subplot.  The subplots that have been grayed out were deemed to have too low of a correlation with the targets to be helpful to the fit.  Note that the subplots showing very discrete structures in the X dimension are the integer-encoded string values.

The columns with the largest affect on the outcome of the targets have been selected, but there is still some additional clearning to be done.  Any rows in the training set that have missing values could throw off the fit, so they will be discarded.

```python
#-- Use only good columns
sub_targets = np.array(df['target'])
sub_data = np.array(df[good_columns])

#-- Create a boolean index of all rows with are all finite values (not Nan, inf)
keep_index = np.all(np.isfinite(sub_data), axis=1)
print "keeping out {} of {} rows.".format(len(np.where(keep_index)[0]), len(sub_data))

sub_targets = sub_targets[keep_index]
sub_data = sub_data[keep_index]
```

---

### Fitting the model

After trying multiple models in the scikit-learn package, the GradientBoostingRegressor model has given the most accurate predictions so far, and has been robust against simply memorizing the dataset.  Therefore, we'll use this model for the final fit and predictions.  

The sections that follow immediately is simply training the model on a subset of the available input data, and using that model to predict and evaluate the remainder of the training dataset.  

```python
#-- Parameters have been chosen to minimize the calculated mean square through an interacitve, 
#-- iterative process.
classifier = ensemble.GradientBoostingRegressor(loss='ls', learning_rate=.07, n_estimators=700)
classifier.fit(sub_data[:3000], sub_targets[:3000])
```

```python
#-- Values come out close, though not exactly the same as the input data.
print classifier.predict(sub_data[3060]), sub_targets[3060]
print classifier.predict(sub_data[4051]), sub_targets[4051]
print classifier.predict(sub_data[3364]), sub_targets[3364]
print classifier.predict(sub_data[3713]), sub_targets[3713]
```

A quick mean_squared error estimate based on the remainder of the training dataset is calculated to be about 1.15.  Doubtless much of the error is due to the noise in the dataset, but the fitting methodology could likely also be improved with a deeper understanding of both the tools and the input datasets.

```python
metrics.mean_squared_error(sub_targets[3000:], classifier.predict(sub_data[3000:]))
```

```python
#-- Re-impliment the classifier using all of the available training datasets to be used in the 
#-- final prediction.
classifier = ensemble.GradientBoostingRegressor(loss='ls', learning_rate=.07, n_estimators=700)
classifier.fit(sub_data, sub_targets)
```

---

### Predicting data

```python
#-- Load the data to be predicted into another dataframe
predict_data = pd.read_table('codetest_2083/codetest_test.txt', delimiter='\t')
```

The test data is loaded in the same was as the training data, with the exception of there being no target column.  Again, string values are encoded into integers, and only the pertinent columns are selected using the good_columns defined in the training dataset.

```python
for colname in predict_data:
    if predict_data[colname].dtype == 'O':
        predict_data[colname] = label_encoder.fit_transform(predict_data[colname])
        
sub_data = predict_data[good_columns]
sub_data = np.array(sub_data.fillna(0))
```

The classifier's predict method is used to evaluate the trained model on the test data, and the results are written to 'predictions.txt'.

```python
predicted_values = classifier.predict(sub_data)
```

```python
with open('predictions.txt', 'w') as outtxt:
    for value in predicted_values:
        outtxt.write('{:.13f}\n'.format(value))
```
